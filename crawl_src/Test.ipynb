```python
import requests
import pandas as pd 
from IPython.display import display, HTML

# Gọi API 
url = "https://restcountries.com/v3.1/all?fields=name,latlng,cca3,cca2,region" 
headers = {"User-Agent": "Mozilla/5.0"} 
response = requests.get(url, headers=headers) 
# Parse JSON 
data = response.json() 
# Chuyển sang DataFrame 
records = [] 
for country in data: 
    if "name" in country and "latlng" in country:
        name = country["name"]["common"] 
        lat, lon = country["latlng"]
        code= country["cca3"] 
        iso_2 = country["cca2"] 
        region= country["region"] 
        records.append({ "country": name, "latitude": lat, "longitude": lon, "code": code, "iso_2":iso_2, "region": region }) 

df_5 = pd.DataFrame(records)
display(HTML(f""" <div style="overflow-x: auto; overflow-y: auto; height:400px; width:100%;"> {df_5.to_html()} </div> """))
```
